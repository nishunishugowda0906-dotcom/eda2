```python
import nltk#nltk stands for natural language tool kit , that helps machine providing some english words to understand the sentiments of the us
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.stem import PorterStemmer#PorterStemmer is a python library that helps the machine to break the  words into their root forms.
from nltk.corpus import stopwords#corpus is like a bag or a group that is consisted with different kinds of english word.
import string#String is that kind of a library in python that helps machine read each and every word  as a token in a sentence.
from nltk.tokenize import word_tokenize
```

```python
stop_words = stopwords.words("english")
ps = PorterStemmer()
```

```python
def preprocessing(text):
  text = text.lower()
  token = nltk.word_tokenize(text)
  new = []
  for i in token:
    if i not in stop_words and i not in string.punctuation:
      stemmer = ps.stem(i)
      new.append(stemmer)
  return" ".join(new)
```

```python
df["clean"] = df["text"].apply(preprocessing)
```

```python
df["clean"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vector = TfidfVectorizer()
X = vector.fit_transform(df["clean"])
y = df["label"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=55)
```

```python
log=LogisticRegression()
log.fit(X_train, y_train)
```

```python
print("train accuracy",log.score(X_train,y_train))
print("test accuracy",log.score(X_test,y_test))
```

```python
def pred_text(text):
    text = text.lower()
    text = preprocessing(text)
    vec = vector.transform([text])
    predict = log.predict(vec)[0]
    if predict == 1:
        return "positive "
    else:
        return "negative"
```

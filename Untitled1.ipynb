```python
# ================= ONE PAGE EDA : GLOBAL SUPERSTORE (COLAB) =================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_excel("/content/global_superstore_2016.xlsx")

# Basic overview
print("Dataset Shape:", df.shape)
print("\nDataset Info:")
df.info()

print("\nMissing Values (%):")
print(df.isnull().mean() * 100)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Convert date columns
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Summary statistics
print("\nNumerical Summary:")
print(df.describe())

# Feature engineering
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month

# Group insights
print("\nSales by Category:")
print(df.groupby('Category')['Sales'].sum())

print("\nProfit by Region:")
print(df.groupby('Region')['Profit'].sum())

# Visualizations
plt.figure()
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title("Sales Distribution")
plt.show()

plt.figure()
sns.boxplot(x='Category', y='Sales', data=df)
plt.title("Sales by Category")
plt.show()

plt.figure()
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Sales vs Profit")
plt.show()

plt.figure()
df.groupby('Year')['Sales'].sum().plot(marker='o')
plt.title("Yearly Sales Trend")
plt.ylabel("Total Sales")
plt.show()

plt.figure()
sns.heatmap(df[['Sales','Profit','Discount','Quantity']].corr(),
            annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# ========================= END OF EDA ======================================
```
